```python
import requests
from google.colab import userdata

# Retrieve the secret key safely
try:
    API_KEY = userdata.get('OPENWEATHER_API_KEY')
except:
    print("⚠️ Please add your 'OPENWEATHER_API_KEY' to the Secrets (key icon) tab!")

# @title 🌦️ Weather Finder
city = "Bengaluru" # @param {type:"string"}
units = "metric" # @param ["metric", "imperial"]

def fetch_weather(city_name):
    url = "http://api.openweathermap.org/data/2.5/weather"
    params = {
        "q": city_name,
        "appid": API_KEY,
        "units": units
    }

    response = requests.get(url, params=params)

    if response.status_code == 200:
        data = response.json()
        temp_unit = "°C" if units == "metric" else "°F"

        print("-" * 30)
        print(f"📍 City: {data['name']}, {data['sys']['country']}")
        print(f"🌡️  Temp: {data['main']['temp']}{temp_unit}")
        print(f"💧 Humidity: {data['main']['humidity']}%")
        print(f"☁️  Sky: {data['weather'][0]['description'].title()}")
        print("-" * 30)
    else:
        print(f"❌ Error: {response.json().get('message', 'Could not fetch data')}")

if API_KEY:
    fetch_weather(city)
```
